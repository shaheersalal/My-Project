Point # 1:
- What is the Net vs Gross Revenue Trend over Time? \
formula: NetRevenue%= (NetSale/GrossSale)*100 

trend includes:
- Monthly/Quarterly/Yearly Revenue Comparison
- Seasonal Trends


```python
df[df["BI Status"]=="Net"].groupby("BI Status")["grand_total"].sum()
```

```python
df[df["BI Status"]=="Gross"].groupby("BI Status")["grand_total"].sum()
```

```python
1175372580.2105 /2910985585.189 * 100
```

Net Revenue is 40%

Trend of Monthly Sale:

```python
a= df[df["BI Status"]=="Net"].groupby("Month")["grand_total"].sum()
b= df[df["BI Status"]=="Gross"].groupby("Month")["grand_total"].sum()

fig,ax= plt.subplots(1,2, figsize=(10,6), sharex= True)
ax[0].plot(a.index,a.values)
ax[0].bin=12
ax[0].set_title("Net")
ax[1].plot(a.index,b.values)
ax[1].bin=12
ax[1].set_title("Gross")
plt.show()
```

Trend of Yaerly Sale:

```python
c= df[df["BI Status"]=="Net"].groupby("Year")["grand_total"].sum()
d= df[df["BI Status"]=="Gross"].groupby("Year")["grand_total"].sum()

fig,ax= plt.subplots(1,2, figsize=(15,6))
ax[0].plot(c.index,c.values)
ax[0].bin=12
ax[0].set_title("Net")
ax[1].plot(d.index,d.values)
ax[1].bin=12
ax[1].set_title("Gross")
plt.show()
```

Analysing Points # 2 \
Which Product Categories or SKUs Generate the Most Net Profit?

```python
category, sku = df[df["BI Status"]=="Net"].groupby(["category_name_1","sku"])["grand_total"].sum().sort_values(ascending=False).index[0]
print(f"{category} is the category and SKU: {sku} generating the most Profit")
```

Analysing Point # 5 
- Which Month or Years Delivers the Highest Average Market Value per Order?

```python
df[" MV "] = df[" MV "].replace(r"[^\d]", "", regex= True)
df[" MV "] = pd.to_numeric(df[" MV "])
df.rename(columns={" MV ": "Market_Value"}, inplace=True)
```

Needed to alter the feature

```python
e= df[df["status"]=="complete"].groupby("Month")["Market_Value"].mean()
f= df[df["status"]=="complete"].groupby("Year")["Market_Value"].mean()

fig,ax= plt.subplots(1,2, figsize=(15,6))
ax[0].plot(e.index,e.values)
ax[0].bin=12
ax[0].set_title("Monthly")
ax[1].plot(f.index,f.values)
ax[1].bin=12
ax[1].set_title("Yearly")
plt.show()
```
